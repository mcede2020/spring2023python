```python
from PIL import Image
import numpy as np
```

```python
img = Image.open('horse.jpeg')#.resize(())
img
```

```python
x = np.array(img)
x.shape
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
Image.fromarray(np.asarray(rgb2gray(x), dtype=np.uint8))
```
