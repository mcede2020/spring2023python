EXERCICE NUMBER 1

```python
import numpy as np
import numpy.linalg
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

#    
def orig_func(n):
    x = np.linspace(1, 15, n)
    plt.grid()
    plt.title(' ')
    plt.plot(x, f(x))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

def approx(n):
    dots = np.linspace(1, 15, n)

    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots)

    w = np.linalg.solve(a, b)  #    
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # 
    
    return res, b


orig_func(250)

n = 5
res = approx(n)
plt.grid()
plt.title(f'   n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
n = 15
res = approx(n)
plt.grid()
plt.title(f'   n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy.random

def approx(n):
    dots = np.linspace(1, 15, n)

    a = np.array([[x**j for j in range(n)] for x in dots])        
    b = f(dots)
    for i in range(len(b)):
        b[i] += np.random.normal(0, 0.01)

    w = np.linalg.solve(a, b)  #    
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  # 
    
    return res, b


orig_func(250)

n = 5
res = approx(n)
plt.grid()
plt.title(f'     n = {n}')
plt.plot(np.linspace(1, 15, n), res[0])
plt.plot(np.linspace(1, 15, n), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
N = 100
def least_squares(N, n):
    dots = np.linspace(1, 15, N)
    a = np.array([[x**j for j in range(n)] for x in dots])
    b = f(dots)
    for i in range(len(b)):
        b[i] += np.random.normal(0, 0.1)

    w, resid, rank, s = numpy.linalg.lstsq(a, b, rcond=None)  #   
    res = np.array([sum([w[j] * x**j for j in range(n)]) for x in dots])  #  
    
    return res, b


orig_func(250)

n = 4
res = least_squares(N, n)
plt.grid()
plt.title(f'    n = {n}')
plt.plot(np.linspace(1, 15, N), res[0])
plt.plot(np.linspace(1, 15, N), res[1], '.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
