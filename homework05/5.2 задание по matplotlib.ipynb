2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x1 = np.linspace(-1, 1, 250)
y1 = x1**2 * np.sin(1/(x1**2)) + x1

plt.title(1)
plt.grid()
plt.plot(x1, y1)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
plt.title(2)
x2 = np.linspace(5, 25, 250)
y2 = abs(1 - (x2 ** 2)/(1 + x2 ** 2))
plt.grid()
plt.semilogy(x2, y2)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
from scipy.special import gamma, psi

plt.figure(figsize=(10, 4))
plt.title(3)
x3 = np.linspace(-5, 5, 250)
y3 = gamma(x3)
x4 = x3
y4 = psi(x4)

plt.grid()
plt.plot(x3, y3, ':r', label='gamma function')
plt.plot(x4, y4, ':g', label='digamma function')
plt.legend()
plt.show()
```

```python
x = np.arange(-np.pi, np.pi, 0.1)
y = x
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)

plt.imshow(Z, cmap='coolwarm', extent=[-np.pi, np.pi, -np.pi, np.pi])
plt.colorbar()
plt.title(1)
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = x
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)

plt.imshow(Z, cmap='viridis', extent=[-3, 3, -3, 3])
plt.colorbar()
plt.title(2)
plt.show()
```
